```python
def gather_images(base_dirs):
    all_images = []
    
    for base_dir in base_dirs:
        for root, _, files in os.walk(base_dir):
            for filename in files:
                # Check if the file is an image (optional, you can filter based on file extensions)
                if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(root, filename)
                    all_images.append(image_path)

    return all_images
```

```python
# Directories to gather images from
directories = [
    'dataset_color/train/ayam',
    'dataset_color/test/ayam',
    'dataset_color/validation/ayam'
]


# Gather images
image_paths = gather_images(directories)

# Example of using the gathered image paths (print them)
for img_path in image_paths:
    print(img_path)
```

```python
def copy_images_to_target(image_paths, target_folder):
    # Create the target folder if it doesn't exist
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)

    for img_path in image_paths:
        # Get the file name from the image path
        filename = os.path.basename(img_path)
        
        # Define the target path where the image will be copied
        target_path = os.path.join(target_folder, filename)
        
        # Copy the image to the target folder
        shutil.copy(img_path, target_path)
        print(f"Copied: {img_path} -> {target_path}")

# Directories to gather images from
directories = [
    'dataset_color/train/wortel',
    'dataset_color/test/wortel',
    'dataset_color/validation/wortel'
]

target_folder = 'dataset_color/all_images/wortel'


# Gather images
image_paths = gather_images(directories)

# Copy images to the target folder
copy_images_to_target(image_paths, target_folder)
```

```python
SOURCE_FOLDER = 'dataset/all_images/wortel'
train_folder = SOURCE_FOLDER + "/train"
val_folder = SOURCE_FOLDER + "/validation"
test_folder= SOURCE_FOLDER + "/test"

os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

all_images = [file for file in os.listdir(SOURCE_FOLDER) if file.endswith('.png')]

train_size = int(0.6 * len(all_images))
val_size = int(0.2 * len(all_images))
test_size = len(all_images) - train_size - val_size

train_images = all_images[:train_size]
val_images = all_images[train_size:train_size + val_size]
test_images = all_images[train_size + val_size:]

def move_files(file_list, dest_folder):
    for file in file_list:
        shutil.move(os.path.join(SOURCE_FOLDER, file), os.path.join(dest_folder, file))

move_files(train_images, train_folder)
move_files(val_images, val_folder)
move_files(test_images, test_folder)
```
